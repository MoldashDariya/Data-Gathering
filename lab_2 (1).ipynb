```python
null_values = df.isnull().sum().sort_values(ascending=False)
null_values
```

```python
df['Gender'] = df['Gender'].fillna('Unknown')
```

```python
df.isna().sum()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(data=df, x='Gender', y='Salary', palette='Set2')
plt.title('Распределение зарплат по полу')
plt.xlabel('Пол')
plt.ylabel('Зарплата')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

```python
print(df.head())
```

```python
df['Bonus %'] = pd.to_numeric(df['Bonus %'], errors='coerce')
```

```python
print(df['Bonus %'].isnull().sum())
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(data=df, x='Bonus %', hue='Gender', multiple='stack', bins=10, palette='Set2', kde=True)
plt.title('Распределение бонусов по полу')
plt.xlabel('Бонусы (%)')
plt.show()
```

```python
plt.figure(figsize=(6, 4))
sns.countplot(data=df, x="Gender", hue="Senior Management")
plt.title('Распределение статуса менеджмента по полу')
plt.show()
```

```python
plt.figure(figsize=(6, 4))
sns.countplot(data=df, x="Team", hue="Senior Management")
plt.title('Статус старшего менеджмента по командам')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.histplot(data=df, x="Bonus %", bins=30, kde=True, hue="Gender")
plt.title('Распределение бонусов по полу')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.histplot(data=df, x="Salary", bins=30, kde=True, hue="Gender")
plt.title('Распределение зарплат по полу')
plt.show()
```
