```python
# Функция для сокращения текста
def shorten_category_name(name, max_length=15):
    if len(name) > max_length:
        return name[:max_length] + '...'  # Оставляем только первые 15 символов
    return name

# Применение сокращения к столбцу категорий
df['short_category'] = df['category'].apply(shorten_category_name)

# Группируем по сокращенным категориям
average_price_by_category = df.groupby('short_category')['discounted_price'].mean().reset_index()

# Построение графика с улучшениями
plt.figure(figsize=(12, 8))  # Увеличиваем размер графика
sns.barplot(x='short_category', y='discounted_price', data=average_price_by_category, palette='viridis')

plt.title('Average Discounted Price by Product Category', fontsize=16)
plt.xlabel('Product Category', fontsize=14)
plt.ylabel('Average Discounted Price', fontsize=14)

# Вертикальные метки с меньшим шрифтом
plt.xticks(rotation=90, ha='center', fontsize=10)

# Подгонка макета
plt.tight_layout()

plt.show()

```

```python
## Сокращаем названия категорий до 10 символов
df['category'] = df['category'].apply(lambda x: x[:10] if isinstance(x, str) else x)  # Проверка, что это строка

# Группируем по категории и вычисляем среднюю цену
average_price_by_category = df.groupby('category')['discounted_price'].mean().reset_index()

# Построение графика
plt.figure(figsize=(12, 6))  # Увеличим размер графика
sns.barplot(x='category', y='discounted_price', data=average_price_by_category, palette='viridis')

# Заголовки и подписи
plt.title('Average Discounted Price by Category', fontsize=16)
plt.xlabel('Category', fontsize=14)
plt.ylabel('Average Discounted Price', fontsize=14)

# Поворот и сокращение надписей на оси X
plt.xticks(rotation=45, ha='right')

# Показ графика
plt.show()
average_price_by_category
```

```python
# Группируем данные по рейтингу и вычисляем среднее количество отзывов
average_reviews_by_rating = df.groupby('rating')['rating_count'].mean().reset_index()

# Строим график
plt.figure(figsize=(10, 6))
sns.barplot(x='rating', y='rating_count', data=average_reviews_by_rating, palette='coolwarm')

# Заголовки и подписи
plt.title('Average Rating Count by Rating', fontsize=14)
plt.xlabel('Rating', fontsize=12)
plt.ylabel('Average Rating Count', fontsize=12)

# Поворот подписей на оси X для читаемости
plt.xticks(rotation=45)

# Показ графика
plt.tight_layout()
plt.show()
average_reviews_by_rating
```

```python
# Подсчитываем количество товаров по категориям
orders_by_category = df['category'].value_counts().reset_index()

# Переименовываем столбцы для удобства
orders_by_category.columns = ['Category', 'Product_Count']

# Печатаем результат
print(orders_by_category)

# Строим график
plt.figure(figsize=(10, 6))
sns.barplot(x='Category', y='Product_Count', data=orders_by_category, palette='Set2')

# Заголовки и подписи
plt.title('Product Count by Category', fontsize=14)
plt.xlabel('Category', fontsize=12)
plt.ylabel('Product Count', fontsize=12)

# Поворот подписей на оси X для читаемости
plt.xticks(rotation=45)

# Показ графика
plt.tight_layout()
plt.show()
```
